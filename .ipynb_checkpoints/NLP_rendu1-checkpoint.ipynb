```python
import pandas as pd
import numpy as np
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification
from sklearn.metrics import accuracy_score, classification_report,  f1_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
import time
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importation des données et prétraitement

Nous commençons par importer les données et par supprimer la catégorie UNE, qui ne peut pas être distinguée de manière fiable des autres classes. Nous regroupons également les classes FRANCE et SOCIÉTÉ, sémantiquement proches, afin d’améliorer les performances de prédiction pour la classe FRANCE–SOCIÉTÉ résultante, mieux représentée après cette étape de prétraitement.


```python
# Importation des données
url = "https://storage.teklia.com/shared/deepnlp-labs/LeMonde2003_9classes.csv.gz"
df = pd.read_csv(url)

# Nombre d'articles
print(f"Total number of articles: {len(df)}")

# Histogramme des catégories
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='category', order=df['category'].value_counts().index)
plt.title("Distribution des catégories dans *Le Monde* (2003)")
plt.show()
```

```python

# Suppression de la catégorie "UNE"
df = df[df["category"] != "UNE"]

# Fusion des catégories France et société
df["category"] = df["category"].replace(
    {"FRA": "FRANCE_SOCIETE", "SOC": "FRANCE_SOCIETE"}
)
```

#  Zero-Shot classification with Hugging Face

## Modèles et labels

Deux modèles ont été testés:
- XLM-RoBERTa-Large-XNLI 
- mDeBERTa-v3-base-mnli-xnli,

ainsi que trois labels:
- Label 1 
     * "ART": "Culture et arts",
     * "ENT": "Économie et entreprises",
     * "FRANCE_SOCIETE": "France et société",
     * "INT": "International et monde",
     * "SPO": "Sports"
- Label 2:
    * "ART": "culture, arts, livres, cinéma, musique",
    * "ENT": "économie, entreprises, finance, marché du travail",
    * "FRANCE_SOCIETE": "actualité en France, société, politique intérieure, faits divers",
    * "INT": "international, diplomatie, conflits, géopolitique",
    * "SPO": "sport, compétitions, football, tennis, résultats"
- Label 3:
    * "ART": "Culture and arts",
    * "ENT": "Economy and business",
    * "FRANCE_SOCIETE": "Society",
    * "INT": "International",
    * "SPO": "Sports"

## Choix des modèles

Les deux modèles testés, à savoir XLM-RoBERTa-large-XNLI et mDeBERTa-v3-base-mnli-xnli sont des modèles entrainés sur des tâches Natural Language Inference (NLI)  multilingues (avec de bonnes performances en français).
Cet entraînement consiste à déterminer la relation logique entre deux phrases : implication (entailment), contradiction ou neutralité (neutral). Il permet ainsi d’évaluer, sans réentraînement spécifique, la relation entre une hypothèse formulée à partir d’un label (par exemple « cet article appartient à la catégorie … ») et le texte de l’article. Ce principe est au cœur du mécanisme exploité par la pipeline zero-shot-classification de Hugging Face.

Les résultats sont très proches entre les deux modèles. Aussi, nous présenterons que les résultats obtenus avec le second modèle moins coûteux en calcul,  notamment grâce à l’architecture DeBERTa, qui repose sur une séparation explicite des informations de contenu et de position.

```python
# Choix du modèle
MODEL_NAME = "MoritzLaurer/mDeBERTa-v3-base-mnli-xnli"
```

```python
#Labels testés
LABEL_MAPS = {
    1: {  # labels très détaillés
        "ART": "culture, arts, livres, cinéma, musique",
        "ENT": "économie, entreprises, finance, marché du travail",
        "FRANCE_SOCIETE": "actualité en France, société, politique intérieure, faits divers",
        "INT": "international, diplomatie, conflits, géopolitique",
        "SPO": "sport, compétitions, football, tennis, résultats"
    },
    2: {  # compromis 
        "ART": "Culture et arts",
        "ENT": "Économie et entreprises",
        "FRANCE_SOCIETE": "France et société",
        "INT": "International et monde",
        "SPO": "Sports"
    },
    3: {  # labels succincts
       "ART": "Culture et arts",
        "ENT": "Économie et entreprises",
        "FRANCE_SOCIETE": "Société",
        "INT": "International",
        "SPO": "Sports"
    }
}
```

Pour comparer les différents labels, nous commencerons par évaluer notre prédiction sur un test set de 50 articles.

```python
# Configuration globale
N_TEST = 50
N_CHARS = 100
results_list = []
all_class_acc = {}
```

```python
# Préparation de l'échantillon (on isole df_test)
# Les textes sont convertis en chaînes de caractères et tronqués aux N_CHARS premiers caractères
df_test = df.sample(N_TEST, random_state=42).copy()
texts = df_test["text"].astype(str).str.slice(0, N_CHARS).tolist()
```

```python
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME)

clf = pipeline(
    task="zero-shot-classification",
    model=model,
    tokenizer=tokenizer,
    device=-1  # CPU ; mets device=0 si GPU dispo
)

# template en français
hypothesis_template = "Cet article parle de {}."
```

```python
# Test rapide sur une phrase simple avec label2
test_text = "Le PSG a gagné le match hier soir au Parc des Princes."
test_labels = list(LABEL_MAPS[2].values())

result =  clf(test_text, test_labels, hypothesis_template=hypothesis_template)
print(f"Catégorie prédite : {result['labels'][0]} avec un score de {result['scores'][0]:.2f}")
```

```python
for config_id, label_map in LABEL_MAPS.items():
    print(f"Test Configuration {config_id}...")
    candidate_labels = list(label_map.values())
    
    # Préparation des vrais labels pour cette config
    y_true = [label_map[c] for c in df_test["category"].astype(str)]
    
    preds = []

    t0 = time.perf_counter()
    
    for txt in texts:
        out = clf(txt, candidate_labels=candidate_labels, hypothesis_template=hypothesis_template)
        preds.append(out["labels"][0])
    
    t1 = time.perf_counter()
    
    # Métriques
    acc = accuracy_score(y_true, preds)
    f1m = f1_score(y_true, preds, average="macro")
    total_time = t1 - t0
    
    # Stockage pour le tableau des résultats
    results_list.append({
        "Config": f"Label_Map_{config_id}",
        "Accuracy": round(acc, 4),
        "F1-Macro": round(f1m, 4),
        "Modèle": MODEL_NAME.split('/')[-1],
        "Temps Total (s)": round(total_time, 2)
    })

# --- AFFICHAGE DU TABLEAU COMPARATIF DES RESULTATS POUR LES DIFFERENTS LABELS ---
df_results = pd.DataFrame(results_list)
print("\nTableau Comparatif :")
display(df_results)

```

```python
import time
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score

batch_size = 8  # à ajuster (GPU: 32/64/128, CPU: 8/16/32)

# Prépare une fois les catégories en str pour éviter de refaire astype dans la boucle
categories_str = df_test["category"].astype(str).tolist()

results_list = []

for config_id, label_map in LABEL_MAPS.items():
    print(f"Test Configuration {config_id}...")

    candidate_labels = list(label_map.values())

    # y_true: map sur les catégories
    # (si jamais une catégorie n'est pas dans label_map, ça plantera: c'est bien pour détecter un souci)
    y_true = [label_map[c] for c in categories_str]

    preds = []

    t0 = time.perf_counter()

    # Batch inference
    for i in range(0, len(texts), batch_size):
        batch_texts = texts[i:i + batch_size]

        outputs = clf(
            batch_texts,
            candidate_labels=candidate_labels,
            hypothesis_template=hypothesis_template,
            # multi_label=False,  # optionnel (par défaut False sur zero-shot)
        )

        # Si batch_texts est une liste, outputs est une liste de dicts
        preds.extend([o["labels"][0] for o in outputs])

    t1 = time.perf_counter()

    acc = accuracy_score(y_true, preds)
    f1m = f1_score(y_true, preds, average="macro")

    results_list.append({
        "Config": f"Label_Map_{config_id}",
        "Accuracy": round(acc, 4),
        "F1-Macro": round(f1m, 4),
        "Modèle": MODEL_NAME.split("/")[-1],
        "Temps Total (s)": round(t1 - t0, 2),
        "Batch size": batch_size,
    })

df_results = pd.DataFrame(results_list)
print("\nTableau Comparatif :")
display(df_results)
```

Le choix des labels joue un rôle déterminant dans les performances du modèle. Contrairement à une approche supervisée classique, les labels ne sont pas de simples codes, mais de véritables descriptions en langage naturel que le modèle utilise pour interpréter le contenu des textes. Des labels trop courts ou trop génériques apportent peu d’information et peuvent conduire à des confusions entre catégories. À l’inverse, des labels trop détaillés ajoutent du bruit sémantique et rendent la décision moins fiable. Les meilleurs résultats sont obtenus avec des labels équilibrés, à savoir:

    - ART → Culture et arts  
    - ENT → Économie et entreprises  
    - FRANCE_SOCIETE → France et société  
    - INT → International et monde  
    - SPO → Sports  
    
Ces labels sontsuffisamment explicites pour caractériser les catégories, tout en restant simples et bien différenciés. 

Nous évaluons à présent les performances du modèle mDeBERTa-v3-base-mnli-xnl, avec les labels sélectionnés précédemment sur un échantillon de 100 articles.

```python
# Utilisation du label 3
label_map = LABEL_MAPS[3]
candidate_labels = list(label_map.values())

# Configuration
N_TEST = 100
N_CHARS = 150
BATCH_SIZE = 2 

# Prédictions  par BATCH 
t0 = time.perf_counter()

outs = clf(
    texts,
    candidate_labels=candidate_labels,
    hypothesis_template=hypothesis_template,
    batch_size=BATCH_SIZE
)

preds = [o["labels"][0] for o in outs]

t1 = time.perf_counter()

# Evaluation du modèle
acc = accuracy_score(y_true, preds)
f1m = f1_score(y_true, preds, average="macro")
total_time = t1 - t0

print(f"Zero-shot results on {len(texts)} articles (first {N_CHARS} chars):")
print(f"  Accuracy:  {acc:.4f}")
print(f"  F1 macro:  {f1m:.4f}")
print(f"  Total time: {total_time:.1f}s")
print(f"  Avg/article: {total_time/len(texts):.2f}s\n")

print("Classification report:")
print(classification_report(y_true, preds, zero_division=0))

# Taille du modèle
n_params = sum(p.numel() for p in model.parameters())
n_params_m = round(n_params / 1e6, 1)

model_size_mb = round(
    sum(p.numel() * p.element_size() for p in model.parameters()) / (1024 ** 2),
    1
)


# Affichage des résultats
df_results = pd.DataFrame([{
    "Config": "Label_Map_2",
    "Accuracy": round(acc, 4),
    "F1-Macro": round(f1m, 4),
    "Modèle": MODEL_NAME.split('/')[-1],
    "Paramètres (M)": n_params_m,
    "Taille modèle (MB)": model_size_mb,
    "Temps Total (s)": round(total_time, 2),
    "Temps/article (s)": round(total_time/len(texts), 2),
    "N_TEST": len(texts),
    "N_CHARS": N_CHARS,
    "BATCH_SIZE": BATCH_SIZE
}])

display(df_results)
```

Nous remarquons toutefois que la classification basée sur des embeddings de phrases fournissait des performances nettement supérieures à celles du zero-shot learning. Les embeddings offrent une meilleure adaptation au corpus étudié et une séparation plus nette entre les catégories. À l’inverse, les approches zero-shot, bien que très flexibles, reposent sur des hypothèses formulées à partir des labels et ne bénéficient pas d’un apprentissage spécifique à la tâche, ce qui limite leur précision.
